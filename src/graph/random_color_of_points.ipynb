# random_color_of_points
Draw points with random color

```python
import random, math
import matplotlib.pyplot as plt 
import matplotlib as mpl

# create points of data in regular position
Nx = 10
Ny = 10
x = []
y = []

for iy in range(Ny + 1):
    for ix in range(Nx + 1):
        x.append(ix * Nx)
        y.append(iy * Ny)    
    
# draw the points
mpl.rcParams['figure.figsize'] = (6, 6)
mpl.rcParams['axes.grid'] = True
plt.scatter(x, y, s=300) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("random color of points") 
plt.show()
```
